```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats

# przygotowanie danych
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
x = np.linspace(-100, 100, 1000)
data100 = stats.norm.pdf(x, 3, 1)
y = np.linspace(-10000, 10000, 100000)
data10000 = stats.norm.pdf(y, 3, 1)
```

```python
mean_100 = np.mean(data100)
print(mean_100)
mean_10000 = np.mean(data10000)
print(mean_10000)
```

```python
median_100 = np.median(data100)
print(median_100)
median_10000 = np.median(data10000)
print(median_10000)
```

```python
mode_100 = stats.mode(data100)
print(mode_100)
mode_10000 = stats.mode(data10000)
print(mode_10000)
```

```python
q0_100 = np.quantile(data100, 0.0)
q1_100 = np.round(np.quantile(data100, 0.25))
q3_100 = np.round(np.quantile(data100, 0.75))
q4_100 = np.round(np.quantile(data100, 1.0))
print(f'Q0: {q0_100}')
print(f'Q1: {q1_100}')
print(f'Q2: {median_100}')
print(f'Q3: {q3_100}')
print(f'Q4: {q4_100}')
q0_10000 = np.quantile(data10000, 0.0)
q1_10000 = np.round(np.quantile(data10000, 0.25))
q3_10000 = np.round(np.quantile(data10000, 0.75))
q4_10000 = np.round(np.quantile(data10000, 1.0))
print(f'Q0: {q0_100}')
print(f'Q1: {q1_100}')
print(f'Q2: {median_100}')
print(f'Q3: {q3_100}')
print(f'Q4: {q4_100}')
```

```python
plt.hist(data100, bins=10)
plt.axvline(x = mean_100, color='red', label = 'Średnia')
plt.axvline(x = median_100, color='green', label = 'Mediana')
plt.axvline(x = mode_100[0], color='orange', label = 'Moda')
plt.axvline(x = q0_100, color='black', label = 'Min')
plt.axvline(x = q1_100, color='black', label = 'Q1')
plt.axvline(x = q3_100, color='black', label = 'Q3')
plt.axvline(x = q4_100, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(data10000, bins=10)
plt.axvline(x = mean_10000, color='red', label = 'Średnia')
plt.axvline(x = median_10000, color='green', label = 'Mediana')
plt.axvline(x = mode_10000[0], color='orange', label = 'Moda')
plt.axvline(x = q0_10000, color='black', label = 'Min')
plt.axvline(x = q1_10000, color='black', label = 'Q1')
plt.axvline(x = q3_10000, color='black', label = 'Q3')
plt.axvline(x = q4_10000, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```
